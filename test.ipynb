```python
import os

import pandas as pd
from transformers import MBartForConditionalGeneration, MBart50TokenizerFast
```

```python
df = pd.read_csv("../data/aestik_loc_test.csv")
```

```python
df = df.dropna(subset="index")
df["index"] = df["index"].astype(int)
```

```python
df_trans = df[["index", "ENGLISH"]].copy()
df_trans = df_trans.dropna(subset="ENGLISH")
```

```python
for idx, row in df_trans.iterrows():
    print(row["ENGLISH"])
```

```python
original_sentence = "Delete File"

model = MBartForConditionalGeneration.from_pretrained("facebook/mbart-large-50-many-to-many-mmt")
tokenizer = MBart50TokenizerFast.from_pretrained("facebook/mbart-large-50-many-to-many-mmt")

# translate Hindi to French
tokenizer.src_lang = "en_XX"
encoded_hi = tokenizer(original_sentence, return_tensors="pt")
generated_tokens = model.generate(
    **encoded_hi,
    forced_bos_token_id=tokenizer.lang_code_to_id["fr_XX"]
)
tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)
```

```python
def get_lang_code(lang: str) -> str:
    lang_dict = {
        "english": "en_XX",
        "russian": "ru_RU",
        "japanese": "ja_XX",
        "korean": "ko_KR",
        "chinese": "zh_CN",
        "french": "fr_XX",
        "portuguese": "pt_XX",
        "spanish": "es_XX",
        "italian": "it_IT",
        "german": "de_DE",
    }

    return lang_dict[lang]


def translate_item(
    input_text: str,
    input_lang: str = "english",
    output_lang: str = "german",
    model=model,
    tokenizer=tokenizer,
) -> str:
    source_lang = get_lang_code(input_lang)
    target_lang = get_lang_code(output_lang)

    model = MBartForConditionalGeneration.from_pretrained(
        "facebook/mbart-large-50-many-to-many-mmt"
    )
    tokenizer = MBart50TokenizerFast.from_pretrained(
        "facebook/mbart-large-50-many-to-many-mmt"
    )

    # translate source to target language
    tokenizer.src_lang = source_lang
    encoded_source_lang = tokenizer(input_text, return_tensors="pt")
    generated_tokens = model.generate(
        **encoded_source_lang,
        forced_bos_token_id=tokenizer.lang_code_to_id[target_lang]
    )
    result = tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)
    return result
```
